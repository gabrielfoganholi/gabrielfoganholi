```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
dados = yf.download('PETR4.SA', start='2023-01-01', end='2023-12-31')
dados
```

```python
dados.columns = ['Abertura', 'Maximo', 'Minimo', 'Fechamento', 'Fech_Ajust', 'Volume']
```

```python
dados = dados.rename_axis('Data')
dados
```

```python
dados['Fechamento'].plot(figsize=(10,6))
plt.title('Variação do preço por data', fontsize=16)
plt.legend(['Fechamento'])
```

```python
df = dados.head(60).copy()
df['Data'] = df.index
df['Data'] = df['Data'].apply(mdates.date2num)
df
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

fig, ax = plt.subplots(figsize=[15, 8])
width = 0.7

for i in range(len(df)):
    if df['Fechamento'].iloc[i] > df['Abertura'].iloc[i]:
        color = 'green'
    else:
        color = 'red'

    ax.plot([df['Data'].iloc[i], df['Data'].iloc[i]],
            [df['Minimo'].iloc[i], df['Maximo'].iloc[i]],
            color=color,
            linewidth=1)

    height = abs(df['Fechamento'].iloc[i] - df['Abertura'].iloc[i])

    ax.add_patch(plt.Rectangle((df['Data'].iloc[i] - width/2, min(df['Abertura'].iloc[i], df['Fechamento'].iloc[i])),
                               width,
                               height,
                               facecolor=color))

# Calcular médias móveis
df['MA7'] = df['Fechamento'].rolling(window=7).mean()
df['MA14'] = df['Fechamento'].rolling(window=14).mean()

# Plotando as médias móveis
ax.plot(df['Data'], df['MA7'], color='orange', label='Média Móvel 7 Dias')  # Média de 7 dias
ax.plot(df['Data'], df['MA14'], color='yellow', label='Média Móvel 14 Dias')  # Média de 14 dias

# Adicionando legendas para as médias móveis
ax.legend()

# Formatando o eixo x para mostrar as datas
ax.xaxis_date()  # Método xaxis_date() é usado para dizer ao Matplotlib que as datas estão sendo usadas no eixo x
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Adicionando título e rótulos para os eixos x e y
plt.title("Gráfico de Candlestick PETR4.SA com matplotlib")
plt.xlabel("Data")
plt.ylabel("Preço")

# Adicionando uma grade para facilitar a visualização dos valores
plt.grid(True)

# Exibindo o gráfico
plt.show()

```

```python
import plotly.graph_objects as go

# Criar o gráfico de candlestick
fig = go.Figure(data=[go.Candlestick(x=df['Data'],
                open=df['Abertura'],
                high=df['Maximo'],
                low=df['Minimo'],
                close=df['Fechamento'])])

# Adicionar as médias móveis
fig.add_trace(go.Scatter(x=df['Data'], y=df['MA7'], mode='lines', name='Média Móvel 7 Dias', line=dict(color='orange')))
fig.add_trace(go.Scatter(x=df['Data'], y=df['MA14'], mode='lines', name='Média Móvel 14 Dias', line=dict(color='yellow')))

# Adicionar título e rótulos dos eixos
fig.update_layout(title='Gráfico de Candlestick PETR4.SA com Plotly',
                  xaxis_title='Data',
                  yaxis_title='Preço')

# Exibir o gráfico
fig.show()
```

```python
import plotly.graph_objects as go

# Criar o gráfico de candlestick
fig = go.Figure(data=[go.Candlestick(x=df['Data'],
                open=df['Abertura'],
                high=df['Maximo'],
                low=df['Minimo'],
                close=df['Fechamento'])])

# Adicionar as médias móveis
fig.add_trace(go.Scatter(x=df['Data'], y=df['MA7'], mode='lines', name='Média Móvel 7 Dias', line=dict(color='orange')))
fig.add_trace(go.Scatter(x=df['Data'], y=df['MA14'], mode='lines', name='Média Móvel 14 Dias', line=dict(color='yellow')))

# Adicionar o volume transacionado como um gráfico de barras
fig.add_trace(go.Bar(x=df['Data'], y=df['Volume'], name='Volume'))

# Adicionar título e rótulos dos eixos
fig.update_layout(title='Gráfico de Candlestick PETR4.SA com Plotly e Volume',
                  xaxis_title='Data',
                  yaxis_title='Preço/Volume')

# Exibir o gráfico
fig.show()
```

```python
import plotly.graph_objects as go

# Criar o gráfico de candlestick com médias móveis
fig1 = go.Figure()

fig1.add_trace(go.Candlestick(x=df['Data'],
                open=df['Abertura'],
                high=df['Maximo'],
                low=df['Minimo'],
                close=df['Fechamento'], name='Candlestick'))

fig1.add_trace(go.Scatter(x=df['Data'], y=df['MA7'], mode='lines', name='Média Móvel 7 Dias', line=dict(color='orange')))
fig1.add_trace(go.Scatter(x=df['Data'], y=df['MA14'], mode='lines', name='Média Móvel 14 Dias', line=dict(color='yellow')))

fig1.update_layout(title='Gráfico de Candlestick com Médias Móveis',
                  xaxis_title='Data',
                  yaxis_title='Preço')

# Criar o gráfico de barras para o volume
fig2 = go.Figure()

fig2.add_trace(go.Bar(x=df['Data'], y=df['Volume'], name='Volume'))

fig2.update_layout(title='Gráfico de Volume',
                  xaxis_title='Data',
                  yaxis_title='Volume')

# Exibir os gráficos separadamente
fig1.show()
fig2.show()
```
